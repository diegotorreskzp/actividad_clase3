+ Muestra cómo utilizar el método .sort() para ordenar la lista de números de forma ascendente.

```python
numeros = [12, 25, 93, 35, 14, 16]
numeros.append(94)
numeros2 = [90, 39, 13]
numeros.extend(numeros2)
numeros.insert(1,20)
numeros.sort()
print(numeros)
```
