+ A partir de la siguiente lista de números, crear una lista nueva solo si el número es divisible por 5.

```python
lista_num = [18,25,29,32,35,42,50,60,63,100]
nuevalista = []
for numero in lista_num:
    if numero%5==0:
        nuevalista.append(numero)
print(nuevalista)
```
